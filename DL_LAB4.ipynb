```python
# ============================================================================
# Assignment No. 4 - MLP model for MNIST digit and Fashion MNIST classification
# Course: UDSC0732 - Deep Learning Lab
# KIT's College of Engineering (Autonomous), Kolhapur
# ============================================================================

# Import necessary libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# ============================================================================
# PART A: MNIST Digit Classification
# ============================================================================

print("\n" + "="*70)
print("PART A: MNIST DIGIT CLASSIFICATION")
print("="*70)

# ------------------------------------------------------------
# A.1(a) Prepare dataset train_test_split
# ------------------------------------------------------------
print("\n1(a) Loading and preparing MNIST dataset...")

# Load MNIST dataset from keras
(X_train_mnist, y_train_mnist), (X_test_mnist, y_test_mnist) = keras.datasets.mnist.load_data()

# Print dataset information
print(f"Training data shape: {X_train_mnist.shape}")
print(f"Training labels shape: {y_train_mnist.shape}")
print(f"Test data shape: {X_test_mnist.shape}")
print(f"Test labels shape: {y_test_mnist.shape}")

# Create validation set from training data
X_train_mnist, X_val_mnist, y_train_mnist, y_val_mnist = train_test_split(
    X_train_mnist, y_train_mnist, test_size=0.1, random_state=42
)
print(f"After split - Training: {X_train_mnist.shape}, Validation: {X_val_mnist.shape}")

# ------------------------------------------------------------
# A.1(b) Normalize dataset
# ------------------------------------------------------------
print("\n1(b) Normalizing MNIST dataset...")

# Normalize pixel values to be between 0 and 1
X_train_mnist = X_train_mnist.astype('float32') / 255.0
X_val_mnist = X_val_mnist.astype('float32') / 255.0
X_test_mnist = X_test_mnist.astype('float32') / 255.0

# Reshape data if needed (for MLP, we'll flatten in the model)
print(f"Normalized pixel range: [{X_train_mnist.min()}, {X_train_mnist.max()}]")

# ------------------------------------------------------------
# A.1(c) Build Multilayer Perceptron Model
# ------------------------------------------------------------
print("\n1(c) Building MLP model for MNIST...")

# Define the MLP model
mnist_model = keras.Sequential([
    # Flatten the 28x28 images to 784 pixels
    layers.Flatten(input_shape=(28, 28)),

    # First hidden layer with 128 neurons and ReLU activation
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.2),  # Dropout for regularization

    # Second hidden layer with 64 neurons and ReLU activation
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.2),

    # Output layer with 10 neurons (for 10 digits) and softmax activation
    layers.Dense(10, activation='softmax')
])

# Compile the model
mnist_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Display model summary
print("\nModel Summary:")
mnist_model.summary()

# ------------------------------------------------------------
# Train the model
# ------------------------------------------------------------
print("\nTraining MNIST model...")
history_mnist = mnist_model.fit(
    X_train_mnist, y_train_mnist,
    validation_data=(X_val_mnist, y_val_mnist),
    epochs=10,
    batch_size=32,
    verbose=1
)

# ------------------------------------------------------------
# A.1(d) Evaluate Model performance
# ------------------------------------------------------------
print("\n1(d) Evaluating MNIST model performance...")

# Evaluate on test data
test_loss_mnist, test_acc_mnist = mnist_model.evaluate(X_test_mnist, y_test_mnist, verbose=0)
print(f"Test Loss: {test_loss_mnist:.4f}")
print(f"Test Accuracy: {test_acc_mnist:.4f}")

# ------------------------------------------------------------
# A.1(e) Predict for test data and plot predicted image
# ------------------------------------------------------------
print("\n1(e) Making predictions and plotting results...")

# Make predictions
y_pred_probs_mnist = mnist_model.predict(X_test_mnist)
y_pred_mnist = np.argmax(y_pred_probs_mnist, axis=1)

# Plot some predictions
fig, axes = plt.subplots(3, 5, figsize=(12, 8))
fig.suptitle('MNIST Predictions', fontsize=16, fontweight='bold')

for i, ax in enumerate(axes.flat):
    ax.imshow(X_test_mnist[i], cmap='gray')
    true_label = y_test_mnist[i]
    pred_label = y_pred_mnist[i]

    # Set title color based on prediction correctness
    color = 'green' if true_label == pred_label else 'red'
    ax.set_title(f"True: {true_label}\nPred: {pred_label}", color=color)
    ax.axis('off')

plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# A.1(f) Visualize using validation loss, actual loss
# ------------------------------------------------------------
print("\n1(f) Visualizing training and validation loss...")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Plot training & validation loss
ax1.plot(history_mnist.history['loss'], label='Training Loss', linewidth=2)
ax1.plot(history_mnist.history['val_loss'], label='Validation Loss', linewidth=2)
ax1.set_xlabel('Epochs', fontsize=12)
ax1.set_ylabel('Loss', fontsize=12)
ax1.set_title('Training vs Validation Loss', fontsize=14, fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot actual loss values
epochs = range(1, len(history_mnist.history['loss']) + 1)
ax2.bar(epochs, history_mnist.history['loss'], alpha=0.7, label='Training Loss')
ax2.bar(epochs, history_mnist.history['val_loss'], alpha=0.7, label='Validation Loss')
ax2.set_xlabel('Epochs', fontsize=12)
ax2.set_ylabel('Loss', fontsize=12)
ax2.set_title('Actual Loss Values per Epoch', fontsize=14, fontweight='bold')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# A.1(g) Plot dataset values for Actual Accuracy and Validation Accuracy
# ------------------------------------------------------------
print("\n1(g) Plotting training and validation accuracy...")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Plot training & validation accuracy
ax1.plot(history_mnist.history['accuracy'], label='Training Accuracy', linewidth=2)
ax1.plot(history_mnist.history['val_accuracy'], label='Validation Accuracy', linewidth=2)
ax1.set_xlabel('Epochs', fontsize=12)
ax1.set_ylabel('Accuracy', fontsize=12)
ax1.set_title('Training vs Validation Accuracy', fontsize=14, fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_ylim([0, 1])

# Plot actual accuracy values
ax2.bar(epochs, history_mnist.history['accuracy'], alpha=0.7, label='Training Accuracy')
ax2.bar(epochs, history_mnist.history['val_accuracy'], alpha=0.7, label='Validation Accuracy')
ax2.set_xlabel('Epochs', fontsize=12)
ax2.set_ylabel('Accuracy', fontsize=12)
ax2.set_title('Actual Accuracy Values per Epoch', fontsize=14, fontweight='bold')
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.set_ylim([0, 1])

plt.tight_layout()
plt.show()

# ============================================================================
# PART B: Fashion MNIST Classification
# ============================================================================

print("\n" + "="*70)
print("PART B: FASHION MNIST CLASSIFICATION")
print("="*70)

# ------------------------------------------------------------
# B.1(c) Prepare dataset train_test_split
# ------------------------------------------------------------
print("\n1(c) Loading and preparing Fashion MNIST dataset...")

# Load Fashion MNIST dataset
(X_train_fashion, y_train_fashion), (X_test_fashion, y_test_fashion) = keras.datasets.fashion_mnist.load_data()

# Class names for Fashion MNIST
fashion_class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
                       'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

print(f"Training data shape: {X_train_fashion.shape}")
print(f"Training labels shape: {y_train_fashion.shape}")
print(f"Test data shape: {X_test_fashion.shape}")
print(f"Test labels shape: {y_test_fashion.shape}")

# Create validation set
X_train_fashion, X_val_fashion, y_train_fashion, y_val_fashion = train_test_split(
    X_train_fashion, y_train_fashion, test_size=0.1, random_state=42
)
print(f"After split - Training: {X_train_fashion.shape}, Validation: {X_val_fashion.shape}")

# ------------------------------------------------------------
# B.1(d) Normalize dataset
# ------------------------------------------------------------
print("\n1(d) Normalizing Fashion MNIST dataset...")

# Normalize pixel values
X_train_fashion = X_train_fashion.astype('float32') / 255.0
X_val_fashion = X_val_fashion.astype('float32') / 255.0
X_test_fashion = X_test_fashion.astype('float32') / 255.0

print(f"Normalized pixel range: [{X_train_fashion.min()}, {X_train_fashion.max()}]")

# ------------------------------------------------------------
# B.1(c) Build Multilayer Perceptron Model
# ------------------------------------------------------------
print("\n1(c) Building MLP model for Fashion MNIST...")

# Define the MLP model for Fashion MNIST
fashion_model = keras.Sequential([
    # Flatten the 28x28 images to 784 pixels
    layers.Flatten(input_shape=(28, 28)),

    # First hidden layer
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.3),

    # Second hidden layer
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),

    # Third hidden layer
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.2),

    # Output layer with 10 neurons
    layers.Dense(10, activation='softmax')
])

# Compile the model
fashion_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Display model summary
print("\nModel Summary:")
fashion_model.summary()

# ------------------------------------------------------------
# Train the model
# ------------------------------------------------------------
print("\nTraining Fashion MNIST model...")
history_fashion = fashion_model.fit(
    X_train_fashion, y_train_fashion,
    validation_data=(X_val_fashion, y_val_fashion),
    epochs=15,
    batch_size=64,
    verbose=1
)

# ------------------------------------------------------------
# B.1(d) Evaluate Model performance
# ------------------------------------------------------------
print("\n1(d) Evaluating Fashion MNIST model performance...")

# Evaluate on test data
test_loss_fashion, test_acc_fashion = fashion_model.evaluate(X_test_fashion, y_test_fashion, verbose=0)
print(f"Test Loss: {test_loss_fashion:.4f}")
print(f"Test Accuracy: {test_acc_fashion:.4f}")

# ------------------------------------------------------------
# B.1(e) Predict for test data and plot predicted image
# ------------------------------------------------------------
print("\n1(e) Making predictions and plotting results...")

# Make predictions
y_pred_probs_fashion = fashion_model.predict(X_test_fashion)
y_pred_fashion = np.argmax(y_pred_probs_fashion, axis=1)

# Plot some predictions
fig, axes = plt.subplots(3, 5, figsize=(15, 9))
fig.suptitle('Fashion MNIST Predictions', fontsize=16, fontweight='bold')

for i, ax in enumerate(axes.flat):
    ax.imshow(X_test_fashion[i], cmap='gray')
    true_label = y_test_fashion[i]
    pred_label = y_pred_fashion[i]

    # Set title color based on prediction correctness
    color = 'green' if true_label == pred_label else 'red'
    true_name = fashion_class_names[true_label]
    pred_name = fashion_class_names[pred_label]

    ax.set_title(f"True: {true_name}\nPred: {pred_name}", color=color, fontsize=9)
    ax.axis('off')

plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# B.1(f) Visualize using validation loss, actual loss
# ------------------------------------------------------------
print("\n1(f) Visualizing training and validation loss...")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Plot training & validation loss
ax1.plot(history_fashion.history['loss'], label='Training Loss', linewidth=2)
ax1.plot(history_fashion.history['val_loss'], label='Validation Loss', linewidth=2)
ax1.set_xlabel('Epochs', fontsize=12)
ax1.set_ylabel('Loss', fontsize=12)
ax1.set_title('Training vs Validation Loss', fontsize=14, fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot actual loss values
epochs_fashion = range(1, len(history_fashion.history['loss']) + 1)
ax2.bar(epochs_fashion, history_fashion.history['loss'], alpha=0.7, label='Training Loss')
ax2.bar(epochs_fashion, history_fashion.history['val_loss'], alpha=0.7, label='Validation Loss')
ax2.set_xlabel('Epochs', fontsize=12)
ax2.set_ylabel('Loss', fontsize=12)
ax2.set_title('Actual Loss Values per Epoch', fontsize=14, fontweight='bold')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# B.1(g) Plot dataset values for Actual Accuracy and Validation Accuracy
# ------------------------------------------------------------
print("\n1(g) Plotting training and validation accuracy...")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Plot training & validation accuracy
ax1.plot(history_fashion.history['accuracy'], label='Training Accuracy', linewidth=2)
ax1.plot(history_fashion.history['val_accuracy'], label='Validation Accuracy', linewidth=2)
ax1.set_xlabel('Epochs', fontsize=12)
ax1.set_ylabel('Accuracy', fontsize=12)
ax1.set_title('Training vs Validation Accuracy', fontsize=14, fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_ylim([0, 1])

# Plot actual accuracy values
ax2.bar(epochs_fashion, history_fashion.history['accuracy'], alpha=0.7, label='Training Accuracy')
ax2.bar(epochs_fashion, history_fashion.history['val_accuracy'], alpha=0.7, label='Validation Accuracy')
ax2.set_xlabel('Epochs', fontsize=12)
ax2.set_ylabel('Accuracy', fontsize=12)
ax2.set_title('Actual Accuracy Values per Epoch', fontsize=14, fontweight='bold')
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.set_ylim([0, 1])

plt.tight_layout()
plt.show()

# ============================================================================
# COMPARISON AND SUMMARY
# ============================================================================

print("\n" + "="*70)
print("COMPARISON SUMMARY")
print("="*70)

print(f"\nMNIST Digit Classification:")
print(f"  - Test Accuracy: {test_acc_mnist:.4f} ({test_acc_mnist*100:.2f}%)")
print(f"  - Test Loss: {test_loss_mnist:.4f}")

print(f"\nFashion MNIST Classification:")
print(f"  - Test Accuracy: {test_acc_fashion:.4f} ({test_acc_fashion*100:.2f}%)")
print(f"  - Test Loss: {test_loss_fashion:.4f}")

print("\n" + "="*70)
print("ASSIGNMENT COMPLETED SUCCESSFULLY")
print("="*70)

# ============================================================================
# CASE STUDY: Autonomous Robot Navigation (Conceptual Implementation)
# ============================================================================
print("\n\n" + "="*70)
print("CASE STUDY: Autonomous Robot Navigation")
print("Domain: Robotics")
print("="*70)

"""
Conceptual Design for Autonomous Robot Navigation MLP:

Inputs (3 features):
1. Distance sensors (multiple readings can be averaged or processed separately)
2. Speed (current velocity)
3. Orientation (current heading/direction)

Outputs (3 classes):
1. Move Forward
2. Turn (Left/Right)
3. Stop

MLP Architecture Concept:
- Input layer: 3 neurons (for the 3 input features)
- Hidden layers: Multiple dense layers with ReLU activation
- Output layer: 3 neurons with softmax activation (for 3 movement decisions)

Training Data Requirements:
- Labeled dataset of sensor readings with corresponding correct movement decisions
- Data should include various scenarios: obstacles, open spaces, different speeds

Implementation Steps:
1. Collect sensor data and corresponding movement decisions
2. Normalize input features
3. Build MLP model
4. Train on collected dataset
5. Test in simulated environment
6. Deploy on physical robot with real-time inference

Note: This is a simplified conceptual design. Real implementation would require
more sophisticated approaches including CNNs for image-based navigation,
reinforcement learning for continuous control, or hybrid models.
"""

print("\nCase Study Summary:")
print("- Problem: Robot movement decision based on sensor inputs")
print("- Inputs: Distance sensors, Speed, Orientation")
print("- Output: Move Forward / Turn / Stop")
print("- Solution Approach: MLP classifier trained on labeled navigation data")
print("- Complexity: Real-world implementation requires additional considerations")
print("  (noise filtering, real-time processing, safety constraints)")
```
